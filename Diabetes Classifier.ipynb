```python
# dropping columns which are not necessary for prediction

to_drop = ['CholCheck','PhysActivity','Fruits','Veggies','HvyAlcoholConsump','AnyHealthcare','NoDocbcCost','Education','Income']

df_1 = df_1.drop(to_drop, axis = 1)
df_1.head()
```

```python
# Plot Correlation Matrix with Heatmap function

plt.figure(figsize = (18, 12))
sns.heatmap(data = df_1.corr(), annot = True, fmt = '.2g', linewidth = 1)
plt.title('Correlation Heatmap', fontsize = 30)
plt.show()
```

```python
# spliting the data into X(Input) and y(Output)

X = df_1.iloc[:,1:]
X
```

```python
# Get the y 
y = df_1.iloc[:,0]
y
```

```python
# splitting data into training set and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
#Using Desision Tree model classification 

from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
#Predicting the Froud from a model by passing test data

y_pred = clf.predict(X_test)
y_pred
```

```python
y_test
```

```python
# Import accuracy_score, confusion_matrix and classification_report from Scikit-Learn

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Calculation of Trained and Test Accuracy

clf_train_acc_DT = accuracy_score(y_train, clf.predict(X_train))
clf_test_acc_DT = accuracy_score(y_test, y_pred)
```

```python
#Printing Accuracy 

print(f"Training accuracy of Desion Tree Classifier is : {clf_train_acc_DT}")
print(f"Test accuracy of Desion Tree Classifier is : {clf_test_acc_DT}")
```

```python
# Printring Confusion matrix of Actual Vs Prediction

print(confusion_matrix(y_test, y_pred))
```

```python
#Printing precision, Recall, F1-Score and Support

print(classification_report(y_test, y_pred))
```

```python
# Using the Random Forest model classification

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
clf = clf.fit(X_train,y_train)
```

```python
#Predicting the Froud from a model by passing test data

y_pred_RF = clf.predict(X_test)
y_pred_RF
```

```python
# Import accuracy_score, confusion_matrix and classification_report from Scikit-Learn

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Calculation of Trained and Test Accuracy

clf_train_acc_RF = accuracy_score(y_train, clf.predict(X_train))
clf_test_acc_RF = accuracy_score(y_test, y_pred_RF)
```

```python
#Printing Accuracy 

print(f"Training accuracy of Desion Tree Classifier is : {clf_train_acc_RF}")
print(f"Test accuracy of Desion Tree Classifier is : {clf_test_acc_RF}")
```

```python
# Printring Confusion matrix of Actual Vs Prediction

print(confusion_matrix(y_test, y_pred_RF))
```

```python
#Printing precision, Recall, F1-Score and Support

print(classification_report(y_test, y_pred_RF))
```

```python
# Using of Neural Network model for Classification

from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='adam', alpha=0.0001,
                    hidden_layer_sizes=(10, 2), random_state=1)
clf.fit(X, y)
```

```python
#Predicting the Froud from a model by passing test data

y_pred_NN = clf.predict(X_test)
y_pred_NN
```

```python
# Import accuracy_score, confusion_matrix and classification_report from Scikit-Learn

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Calculation of Trained and Test Accuracy

clf_train_acc_NN = accuracy_score(y_train, clf.predict(X_train))
clf_test_acc_NN = accuracy_score(y_test, y_pred_NN)
```

```python
#Printing Accuracy 

print(f"Training accuracy of Desion Tree Classifier is : {clf_train_acc_NN}")
print(f"Test accuracy of Desion Tree Classifier is : {clf_test_acc_NN}")
```

```python
# Printring Confusion matrix of Actual Vs Prediction

print(confusion_matrix(y_test, y_pred_NN))
```

```python
#Printing precision, Recall, F1-Score and Support

print(classification_report(y_test, y_pred_NN))
```

```python
#Summarising the all Models' Accuracies

Summary_Table =[[clf_train_acc_DT, clf_train_acc_RF,  clf_train_acc_NN],[clf_test_acc_DT,clf_test_acc_RF,clf_test_acc_NN ]]
Summary_Table
```

```python
#Converting it into Numpy array

Summary_table = np.array(Summary_Table)
Summary_table
```
